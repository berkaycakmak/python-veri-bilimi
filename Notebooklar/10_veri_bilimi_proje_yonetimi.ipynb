# **Veri Bilimi Proje Yönetimi**

## **İş Problemlerine Yaklaşım Geliştirmek**

Algoritmaya veri uydurmak yerine verilere algoritma uydurmak gereklidir.

### İş Problemlerine Yaklaşım Geliştirme Kültürü

* Amaç: Geliri artırmak, süreç optimizasyonu, stratejik kararlar almak, bütçe yönetimi, insan yönetimi.
* İş problemi: Bu amaçları gerçekleştirmenin önündeki engellere iş problemleri denir.
* Yaklaşım: Problemleri çözmek için geliştirilen öneri ve projelerdir.
* Araçlar/Algoritmalar
* Çözüm

## **Veri Bilimi Proje Döngüsü**

## **Bazı Başlangıç Soruları**

* İlgili birimlerin ya da iş yerinin ihtiyacı nedir?
* Bu proje ile hangi problemi çözeceğiz?
* Attığımız taş ürküttüğümüz kuşa değecek mi?
* Bu işin somut çıktısı ve işe katkısı nedir?
* Neyi ölçmek istiyoruz?
* Neyi tahmin etmek istiyoruz?
* Bu modeli ya da tahmini kim kullanacak?
* Projenin başarı metriği nedir?
* Bu işin doğrulanmasından ve kontrolünden kim sorumlu olacak?
* Bu proje ne kadar sürecek?
* Hangi departmanlar hangi seviyelerden işin içinde olacak?

## **Adım 1: Sağlam Bir İlk Adım**

* Problemin Apaçık Tanımlanması
* Proje Beklenti Yönetimi
* Projenin Başarı Metriğinin Ne Olacağının Belirlenmesi

## **Adım 2: Literatür Taraması ve En İyi Uygulamaların Araştırılması**

Başka kişilerin benzer problemlere nasıl yaklaştığını, problemi nasıl aştığını ve benzeri bilgileri inceleyerek ve bu bilgilerden yola çıkarak projeye başlanması ve devamının getirilmesi fayda sağlayacaktır.

## **Adım 3: Veriyi Anlamak**

* Veri tabanlarında iz sürmek
* Verinin oluşma şeklini anlamak
* Verinin hayat döngüsünü anlamak
* Verinin ilgili proje için uygunluğunun anlaşılması

## **Adım 4: Özellik Mühendisliği**

* Tidy Süreci: Tidy data, bir çalışma öncesinde verinin düzenli hale getirilmesidir. Düzenli ifadesi her bir satırın ilgilenilen gözlem birimini, her bir sütunun bir değişkeni ifade etmesi durumudur.
* Keşifçi Veri Analizi: Düzenli veri üzerinde çeşitli keşiflerin yapıldığı bölümdür.
* Feature Engineering: Veri setinde hedefimize yönelik bize katkı sağlayacağını düşündüğümüz değişken üretmek, analiz etmek, değerlendirmek ve modelleme öncesi hazır hale getirmektir.

## **Adım 5: Veri Ön İşleme**

* Veri Temizleme (Data Cleaning/Cleasing)
* Veri Standardizasyonu/ Dönüşümü
* Veri İngirgeme

## **Adım 6: Modelleme ve Model Başarı Değerlendirme**

Elimizdeki tüm modellerin veri üzerinde denenmesi ve tune edilmesi bölümüdür. Tüm modeller test edildikten sonra tahmin başarısı en yüksek modelin seçimi bu adımda yapılır.

## **Adım 7: Çalışmanın Canlı Sistemlere Entegrasyonu**

* İş birimlerini beslemek
* Canlı sistemkeri anlık beslemek (anlık tahminsel modeller çalıştırmak)
* Canlı sistemler için tablolar oluşturmak
